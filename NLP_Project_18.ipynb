```python
#Proprocesing and lemmitization of each document

from wikipedia.util import print_function
import re
from nltk.tokenize import word_tokenize



def preprocess_doc(input_document):

  stop_words = set(stopwords.words('english'))
  lemmatizer = WordNetLemmatizer()
  words = word_tokenize(input_document)
  words = [word.lower() for word in words if word.isalpha()]
  words = [word for word in words if word not in stop_words]
  words = [lemmatizer.lemmatize(word) for word in words]
  result = ' '.join(words)
  return result

nature_preprocessed = preprocess_doc(nature_doc)
sustainability_preprocessed = preprocess_doc(sustainability_doc)
air_pollution_preprocessed = preprocess_doc(pollution_doc)
environment_preprocessed = preprocess_doc(environment_doc)

environment_preprocessed

```

```python
                      #REORGANIZING EACH WEBPAGE IN TERMS OF THE LIST OF ENTITIES (CLICKABLE KEYWORDS EXCEPT FOR REFERENCES) (Part of Task 1)

import requests
from bs4 import BeautifulSoup

def extract_entities_from_wikipedia(page_title):
    # Wikipedia URL
    wikipedia_url = f"https://en.wikipedia.org/wiki/{page_title}"

    # Make an HTTP request to the Wikipedia page
    response = requests.get(wikipedia_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Find all clickable links within the content
        entity_links = soup.find_all("a")

        entities = []

        for link in entity_links:
            entity_text = link.get_text()
            #entity_text = link.get('title')
            entity_href = link.get("href")

            # Check if the link is a Wikipedia internal link                                                                                                                                    entities = [entity for entity in entities if not entity.startswith("doi") and not entity.startswith("PMC") and not entity.startswith("PMID")]
            if entity_text and entity_href and entity_href.startswith("/wiki/") and not any(char.isdigit() for char in entity_text) and not entity_text.startswith("\n"):
                entities.append(entity_text)
        return entities

    else:
      #  continue
      #  print(f"Failed to retrieve the Wikipedia page for '{page_title}' (HTTP Status Code: {response.status_code})")
        return []

def entities_extractor():
    page_title = "Nature"  # Replace with the title of the Wikipedia page you want to extract entities from

    entities = extract_entities_from_wikipedia(page_title)

    if entities:
        print("Clickable Keywords (Entities):")
        for entity in entities:
            print(entity)
    else:
        print("No clickable keywords found on the Wikipedia page.")
x = extract_entities_from_wikipedia("Nature")
print(x)
```

```python
                                    #Task 5: correlation between the semantic similarity result and each of the above Wikipedia based similarity

import nltk
from nltk.corpus import wordnet
from scipy.stats import spearmanr
import pandas as pd

# Download WordNet data
nltk.download("wordnet")

# Define the pairs of words for which you want to calculate similarity
document_names = ["nature", "sustainability", "pollution", "environmentally_friendly"]

# Calculate WordNet-based similarity using Wu and Palmer method
def wup_similarity(word1, word2):
    # Find the synsets for each word
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)

    # Initialize a list to store similarity values
    similarities = []

    for synset1 in synsets1:
        for synset2 in synsets2:
            similarity = synset1.wup_similarity(synset2)
            if similarity is not None:
                similarities.append(similarity)

    # Return the maximum similarity value found
    return max(similarities, default=0)

wordnet_similarities = []

for word1 in document_names:
    row = []
    for word2 in document_names:
        similarity = wup_similarity(word1, word2)
        row.append(similarity)
        # print(similarity)
    wordnet_similarities.append(row)

# Define Wikipedia-based similarities
wikipedia_similarities = cosine_similarity_df
wikipedia_similarities = [[value[0][0] for value in row] for row in wikipedia_similarities.values]
# print(wikipedia_similarities)

# Calculate the correlation between WordNet and Wikipedia similarities
wordnet_values = []
wikipedia_values = []

for i in range(len(document_names)):
    for j in range(i, len(document_names)):
        print(wordnet_similarities[i][j])
        wordnet_values.append(wordnet_similarities[i][j])
        print(wikipedia_similarities[i][j])
        wikipedia_values.append(wikipedia_similarities[i][j])

print(wordnet_values)
print(wikipedia_values)

correlation, _ = spearmanr(wordnet_values, wikipedia_values)

print("WordNet Similarities:")
for i in range(len(document_names)):
    for j in range(i, len(document_names)):
        print(f"{document_names[i]} - {document_names[j]}: {wordnet_similarities[i][j]}")

print("\nWikipedia Similarities:")
for i in range(len(document_names)):
    for j in range(i, len(document_names)):
        print(f"{document_names[i]} - {document_names[j]}: {wikipedia_similarities[i][j]}")

print("\nCorrelation between WordNet and Wikipedia Similarities:", correlation)
```
